# Quick summary
*The actual notes starts from the next heading. This summary is here so I can quickly remember how to do valuaiton after a hiatus*
  1. We usually value the whole company and at the end subract out the debt. This is because using netincome as the basis of valuation is a bad idea because accounting tricks, one-off expenses end up giving us the wrong picture. Using post-tax opinc is better
  2. From the company financials, find out return on invested capital and reinvestment rate. Operating income is expected to grow at the rate of $g = RIR*ROIC$. Refer to [Formulae section](#Formulae) section below for computing RIR and ROIC. 
  3. Project future growth of post-tax op inc based on current post-tax post inc and g for some years (for which extra-ordinary growth is expected)
  4. Calculate FCFF using projected post tax op inc and RIR
  5. Convert all FCFF terms to present values using weighted average cost of capital
      1. Find wacc by finding cost of equity (based on risk free rate, levered beta and risk premium) and cost of debt (based on publicly available bond yields or by adding credit-rating based spread to risk-free rate)
  6. Add present values. Calculate a terminal value based on a fixed growth rate less than or equal to current risk free rate. Steady growth period's cost of capital can be taken as the risk free rate or slightly higher

# Asset
+ An "asset" is anything that generates cash flows at any/several points in the future.
+ An asset can be financed by the asset's "owners", a.k.a., equity holders, by either using their own money or by borrowing from a bank. i.e., Assets = Equity + Liabilities. So a balance sheet is nothing but assets on the left hand side and what financed those assets on the right hand side
+ An asset is said to be operating if it generates a cashflow. Suppose you buy a car with the intention of using it as a taxi, it is an operating asset if you do use it as a taxi and generate revenue. If the car is sitting idle in the parking lot, it is a non-operating asset
+ The value of an asset is the sum of all cashflows (discounted at the appropriate rates), it generates from now to forever in the future. i.e., it's present value. It's past does not matter. An asset itself doesn't have any value unless it generates cash flows. In the previous example, the car sitting in the parking lot was a non-operating asset. Although it has a market value, unless it is sold, it doesn't generate a cash flow, and hence, doesn't have a value. So a car that is going to be sitting in a parking lot forever has a present value of zero dollars, and a net present value (N.P.V.) of negative value
+ The money invested has a cost associated with - the opportunity cost - whether the money comes from owners' own pockets or from creditors (or a combination of both). Both the owner as well as the creditor, by investing money into the asset, are foregoing the opportunity to invest that money in something else, and consequently, foregoing the cashflows that something else will generate. If the asset in question generates more cashflows that the alternative asset, then the investment can be adjudged to be good investment. The cashflow generated by the asset as a percentage of the invested capital is essentially the asset's ROIC - Return on Invested Capital. The same for the alternative investment - i.e., the opportunity cost - is the Cost of Capital (COC). When there is an excess return, i.e., ROIC - COC, the investment is a good investment. 
    + For ex., if you buy a juice maker with 100 dollars borrowed entirely from the bank, say at 5% annual interest, and sell juice and make a profit, then as long as you make more than 5 dollars in profit in a year, your investment would be considered a good investment. Say you make actually 6 dollars of annual profit, then 6%-5% = 1% is your company's excess return. Excess return is nothing but the value you created - If you generated only a 5 dollar annual profit profit, then you would have give away all of it to the bank and you will back where you were an year ago. So you would not have created any value. In other words, value is created when you did something with the money that bank could thought would earn only 5%, considering the risk of the investment (using whatever objective or subjective method of assessing), and you could make more than 5%.
    + One was of thinking about COC and ROIC is through DCF. In the juice company example, if you generated only 5 dollars of annual profit in a year after investing 100 dollards, then value of the company would be 100 dollars as well: So you haven't creating any value in a year of operating the company
    $$
    PV = \dfrac{100\ dollars\ +\ 5\ dollars}{1+0.05} = 100\ dollars
    $$

    
+ An asset's value must be independent of how one uses it. For instance, the value of a house should be independent of whether you are going to live in it or rent it out. In fact, that is exactly how one can evaluate the objective value of the pleasure of owning a house - the opportunity cost of those pleasures is what you are giving up by having that pleasure - and it is the cashflows you could have got by renting out that house. 
    + Another way to think is, suppose you buy a house, say house A, and rent an equivalent house, house B. It means, you can collect \\$x per month as rent from house A and simply use to pay your rent - Since both houses are similar, the rents must be identical. In this case, from the point you purchase house A, you get to enjoy house B without spending one extra dollar more. So house B must have the same price as house A. The pleasure of enjoying house B can objectively be defined as the cash flow generated by house A discounted by appropriate rates. So it makes sense to always evaluate an asset's value from the point of view of cash flow it would generate, whether or not you are going to use to asset for a business or for personal use
    + The above example can be twisted as saying that, if you want to evaluate house A, look at the cash flows you will be saving by living in house A - i.e., not paying rent on an equivalent house. So house A price must be equal to all the rent saved discounted by appropriate rates.
    + There is a nuance related to comparing one living in one's property vs. renting it out to others - if a property is used for income generating activity (such as in the latter), in some countries, the property's owners can deduct depreciation from the rental income before it is taxed. 
+ The money spent on creating an asset is a "capital expense". The other type of expense is an operating expense. The easy way to judge if something is capex or opex, is to check whether it is going to generate cash. In other words, money spent as capital expense comes back to you (hopefully at a higher value), but money spent on an operating expense is forever lost.
    - A house is a capex as it can generate cashflows from rent and then from an eventual sale. So the amount you paid to buy the house is a capex. But minimum energy costs you have to pay, whether you are using it or not, the maintainence costs you have to spend whether you are using it or not, are all opex. Note that, when valuing a house for purchase, you will consider cashflows coming to your pocket - i.e., if it is rent, then you would subtract any tax paid on it, both income and house tax, maintenance expenses etc., and then use the remainder as the cashflow. In other words, opex is already accounted for while valuing the house for purchase.
    - Educating expense are a capex - you expect to use the education to get a job and generate a cash flow from the salaries. Food is an operating expense - you will never recover anything from it. But a person needs food to operate. Hence it is called an operating expense. Operating expense is whatever is essential of keeping the asset operating - preventing it from becoming a non-operating asset
    - COGS appear only if goods are created. OpEx is the expense that is independent of whether or not the goods are created. For example, if you are a copacker, electricity cost incurred when the machines are running - doing packaging - is part of COGS. Suppose you get no orders for a month, COGS will correspondingly be zero as you will simply switch the machines off. But just to keep the business operational, you have to spend a minimum amount to keep lights on at the front office, pay lobby person, phone bill, pay accounted to file IT returns etc. They will be part of OpEx.
+ Most assets depreciate in value over time. A car purchased for use as a taxi, will be losing its value over time. This depreciation needs to be accounted as an operating expense. 
    - Right at the moment you purchase a car, your assets remain unchanged in value - only changed in form: What was earlier in the form of cash is now in the form of a car. But from that point on, the car's value starts depreciating and your asset value starts going down. If the car doesn't generate cashflows to makeup for the depreciation (or more than make up for it), then it means you have destroyed value. 
    - If you purchase a car at say Rs. 10 lakhs, and if you try to sell it a year from now, if you get Rs. 8 lakhs, it means the value of the car depreciated by Rs. 2 lakhs. You need to account that as an operating expense. Assets can decrease linearly (straight line depreciation), i.e., Rs. 2 lakhs every year, or non-linearly. Cars usually depreciate non-linearly. But most corporate assets are regarded as depreciating in a straight line. (Considering depreciation as an opex is probably only for accounting purposes because doing so will allow you to save on tax. Furthermore, you need to account for the lost value somewhere in the balance sheet, failing which, the balance sheet won't be balanced! However, conceptually capex and opex are two disjoint ideas)
+ Any capex accounted as opex over a period of time is said to be "amortized". Depreciation is one type of amortization. But usually the word "amortizing" is used while talking about capex that are not assets - Money spent buying a truck is "depreciated", whereas, license paid to operate the trucking company is "amortized". The basic process of depreciation and amortization are same. hence they are always referred to together 
+ The part of the assets remaining after subtracting liabilities is what belongs to the equity holders. The moneys borrowed to finance an asset is called "leverage". If, say, you buy a \\$500K house with $100K of your own money and \\$400K of banks money, then it is said to be a leveraged buyout or "LBO", and you are levered 5:1. This is because, if the value of the house appreciates to, say, \\$600K (after netting out interest expenses), then you would have made \\$200K, that is twice the money: You contributed only 1/5th or 20\% for acquiring an asset, but made out a 5x profit of 100\%. This is because the bank gets only \\$400K no matter how much the house appreciates. It is called "leverage" because it acts like a lever in physics. Leverage works the other way as well. If the house's market value reduces by 20\% the equity holder loses 100\% of his investment.
+ Borrowing might make good sense if one can invest that money at the same risk level as the bank investing it is taking, while getting more returns than the bank interest rate. In other words, you have a talent/knowledge that the bank doesn't have - and you use this talent/knowledge to achieve returns that people, on an average, cannot achieve by taking the same amount of risk

# Liabilities
+ Debt doesn't necessarily mean short-term/long-term loans from a bank. A long-term lease on a building is also a debt. Debt has the following characteristics:
    - It has contractual commitments
    - It is tax deductible 
    - If contractual obligations are not met, some bad things (ex. foreclosure) happen
+ Interest payments, in many countries, are tax deductible. So financing using debt is attractive from that angle
+ Unlike individuals, most companies don't make regular payments (EMI) towards both the capital and interest. They pay only interests until the maturity date. And at the maturity date, they take another loan and pay off the old loans capital amount with the new loan, and then start paying interests towards the new loan. This is done to avoid extraordinary expenses in any given quarter

# Income and expenses
+ Revenue can go up because of one or more of the following reasons (mentioned here for product companies. But a analogue works for service companies too)
    - More units sold
    - Same units sold as before, but sold at higher prices. This is not in itself bad. This means that customers saw increased value in the product
    - Inflation increased. In this case, the real revenue didn't go up - only the nominal revenue went up
+ Gross income = Revenue - Cost of Sales (COGS). COGS is like the raw material needed to make a product. Gross income hence represents the value added by the product. If we add all the value additions, i.e., the gross incomes, of all the companies in a country, then it becomes the Gross Domestic Product. In other words, if companies in India, small and big, producing every imaginable good/service, were to merge together, then the Gross income of that super company would be the GDP of India. In that case, since, all the raw material coming from within India would be also be part of the company's value addition, it will be considered as part of the Gross Income. Only the raw materials coming from outside the country - in this case, outside the super company - will be considered as COGS. This is basically all the imports.
+ One can broadly classify expenses into four types: Gross expenses (COGS), operational expenses (SG&A, office energy costs etc.), capital expenses (D&A) and financial expenses (interest expense)
+ COGS is that expense that linearly increases with units sold. It is that expense is that directly associated with the production of goods. For ex., energy cost in running production machinery is COGS, but energy cost for the office building that houses marketing, sales teams is opex
+ R&D expenses are treated as operating expenses for accounting purposes. But these are really capex. 
+ Net income could increase because:
    - Revenue increased, which could mean the company now has more customers or the value of the company's goods/services has increase or both
    - COGS decreased. This may be a cyclical thing which would hence be temporary. If, somehow, a new innovation has reduced COGS, this could be a permanent thing, which is very good for your company
    - Company optimized opex. While this is good, the thing to remember is, the net income increase in this case doesn't mean the company is growing. It only became more efficient. In fact, it is possible that losing market share may be offset by temporary improvement in opex (say, by firing some staff), which may mask the company's problems in the short term, but exacerbate them in the long term
    - Long term lease is often listed in the income statement as opex. But due to its nature, it is actually debt. So one should be careful with it

# Special type of companies
## Real estate investment trust
+ Since most of real estate is about converting assets from one form to another (cash to property and back to cash), income statement may not reflect how much the firm is growing - Retained income may be small or zero, but assets may be growing enormously every year 

## Banks
+ By design, they aren't supposed to grow fast
+ Liabilities are bank savings deposits, fixed deposits etc. Non-cash assets are mostly loans, fixed income securities. Corresponding, income will chiefly be interest income and expenses, interest expenses

# Valuation
+ Suppose you are planning to buy a taxi company that has 4 taxis, one office building in assets and no liabilities
    + The balance sheet would show the assets and their depreciation. Suppose they had chosen to use straight line depreciation, considering that, we will get a **book value** of the assets
    + The market may view depreciation of cars in a non-straight line manner. So if one were to buy equivalently old building, cars in the market, one has to pay the **market value** for the assets. This is the "fair" value.
    + The **intrinsic value** of the company is the cash flows their assets generate. The predicted net incomes for all the times to come plus the cash one would get eventually by selling unusable cars (for scraps, for ex) and the building, discounted by appropriate rates should give the intrinsic value.

# Valuation course notes (Aswath Damodaran)
[Playlist link](https://www.youtube.com/playlist?list=PLUkh9m2BorqnKWu0g5ZUps_CbQ-JGtbI9)

To evaluate the present value of any asset, we need cashflows (expected to be) generated by the asset, the associated timeline and the discount rates we need to use on each of the cashflow. i.e., 

$$
PV = \dfrac{C_1}{1+r_c} + \dfrac{C_2}{(1+r_c)^2} + \dfrac{C_3}{(1+r_c)^3} + \ldots  
$$

Here, $C_{i}$ are cashflows for period $i$ (usually periods are years) and $r_c$ is the (opportunity) cost of capital that was used to create the asset. If the present value works out to be the same as the capital used to create the asset, then you didn't create any value, and didn't destroy any value. You wasted your time. 

A company is just like any other asset. No one can predict future cash flows of a company. And according to Aswath Damodaran, past cashflows are not a good indicator of future cash flows. So, the only true cash flow that we can base our projections of future cashflows on, is the most recent cashflow $C_{0}$. One can assume that this cashflow grows at a rate of $g$ for the future years. Considering that, and the fact that the future could extend all the way to $\infty$, our $PV$ calculation becomes

$$
\begin{align}
PV &= \dfrac{C_0(1+g)}{1+r_c} + \dfrac{C_0(1+g)^2}{(1+r_c)^2} + \ldots + \dfrac{C_0(1+g)^\infty}{(1+r_c)^\infty} \\
   &= \dfrac{C_0(1+g)}{r_c-g} \left[ 1 - \left(\dfrac{1+g}{1+r_c}\right)^\infty \right] 
\end{align}
$$

However, the above formula has two problems: 
    1. Although, in theory, it is possible for a company to grow its cashflow without reinvesting some money from previous cashflows, in practice, some amount of reinvestment is required. That part is lost to the company. This changes the $PV$ calculations as now we have to mutliply every term in the summation by a $(1-rir)$ factor, where $rir$ is the reinvestment rate. So we get,
    
$$
\begin{align}
PV &= \dfrac{C_{0}(1+g)(1-rir)}{1+r_c} + \dfrac{C_{0}(1+g)^2(1-rir)}{(1+r_c)^2} + \ldots + \dfrac{C_{0}(1+g)^\infty(1-rir)}{(1+r_c)^\infty} \\
   &= \dfrac{C_{0}(1-rir)(1+g)}{r_c-g} \left[ 1 - \left(\dfrac{1+g}{1+r_c}\right)^\infty \right] \\
   &= \dfrac{FCF_{0}(1+g)}{r_c-g} \left[ 1 - \left(\dfrac{1+g}{1+r_c}\right)^\infty \right]
\end{align}
$$

where, $FCF_0 = C_0 (1-rir)$ is the most recent "free cash flow" to the company.
    2. The other problem is that, the $PV$ will become $\infty$ when $g > r_c$. So only when we are evaluating mature companies that have reached a stage where their growth is slower than the cost of capital. So what do we do for fast growing companies? The tweak used by the professor is that, we split the companies growth periods into two phases: Phase1, in which the company is growth at a fast rate $g1>r_c$ for the first few years and then $g2<r_c$ for all the years after that. He also allows for different reinvestment rates for the two phases. So we will get the following DCFs for the two phases:

$$
\text{Phase 1 DCF: } \dfrac{C_0(1+g1)(1-rir1)}{1+r_c} + \dfrac{C_0(1+g1)^2(1-rir1)}{(1+r_c)^2} + \ldots + \dfrac{C_0(1+g1)^N(1-rir1)}{(1+r_c)^N} \\
\text{Phase 2 DCF: } \dfrac{C_N(1+g2)(1-rir2)}{(1+r_c)^{N+1}} + \dfrac{C_N(1+g2)^2(1-rir2)}{(1+r_c)^{N+2}} + \dfrac{C_N(1+g2)^3(1-rir2)}{(1+r_c)^{N+3}} + \ldots
$$

Phase 2 goes on to infinity. So it is called the "Terminal value". In terms of free cash flows, the DCFs can be rewritten as:

$$
\text{Phase 1 DCF: } C_0(1-rir1) \sum_{i=1}^{N} \dfrac{(1+g1)^i}{(1+r_c)^i} = \dfrac{FCF_0(1+g1)}{r_c-g1} \left[ 1 - \left(\dfrac{1+g1}{1+r_c}\right)^N \right] \\
\text{Phase 2 DCF: } \dfrac{C_N(1+g2)(1-rir2)}{(1+r_c)^{N+1}} \sum_{j=0}^\infty \dfrac{(1+g2)^j}{(1+r_c)^j}  = \dfrac{FCF_{N+1}}{(1+r_c)^N(r_c - g2)}
$$

where, $FCF_{N+1}$ is the free cash flow of the $(N+1)^{th}$ period, and is equal to $ C_N (1+g2)(1-rir2) $, and $C_N = C_0(1+g1)^N$  (If this is confusing, remember that the cashflow kept growing at the pace of $g1$ for N years. On $n+1$th year, it would have grown at the new pace of $g2$ and the reinvestment would have chanced to $rir2$. There is a bit of arbritrariness here: We have argued that growth is partly the result of reinvestment. So you could say that a new growth rate of $g2$ could have only resulted from a new reinvestment rate of $rir2$. So the $N$th year FCF should really be $C_0(1+g1)^N(1-rir2)$. But that would make it difficult to come up with closed form representation and wouldn't make much of a difference if $rir1$ isn't that different from $rir2$.). Finally, we can get $PV$ by summing DCF of both phases:

\begin{equation}
PV = \dfrac{FCF_0(1+g1)}{r_c-g1} \left[ 1 - \left(\dfrac{1+g1}{1+r_c}\right)^N \right] + \dfrac{FCF_{N+1}}{(1+r_c)^N(r_c - g2)}
\label{eq:PV} \tag{1}
\end{equation}

Professor Damodaran's book/videos are all about finding the unknowns in the above equation. He explains many nuances along the way. But the bottom line is that evaluating the intrinsic value of the company is all about working out the above equation

*Note^: Professor uses $rir2$ sometimes while calculating $N^{th}$ year $FCF$. But if we do that, the formula doesn't render into a compact form. Although our result will be different, there are so many other variables that may be different from us and the professor - $N$ itself is subjective. So is $rir2$. One could argue that the cost of capital would be lower when the company is in phase 2, and so much more. Apart from all these, it is pretty unrealistic that the company will move over from phase 1 to phase 2 abruptly, or that there will only be two phases. So everything discussed above should be considered as laying conceptual foundations for DCF based valuation. The actual valuation would have to factor in firm specific characteristics and one may have to make many adjustments to the above DCF calculations to come up with a valuation.

## Cashflows for valuation
In equation $\eqref{eq:PV}$, we mentioned cashflows and free cash flows. But what are they? In this section we explore this question.

Imagine a person starting a simple business - say, a typical Indian barber shop. Suppose the person uses his own money to buy a small property, buy chairs and other hair cutting, cleaning equipment - i.e., he didn't get any loan. Say he employs two barbers. He pays their salary, energy costs and for raw materials such as razon blades, shampoo etc. Assuming this is a profitable business, every year, after accounting for all expenses, income tax etc., there will be some money (net income) left. This is the cashflow to the company for that year. Now the owner may choose the following options, or combinations thereof:

1. Keep the cashflow in the company's bank account as cash
2. Transfer it to his personal account. This is basically dividend.
3. Reinvest the money to create new operating assets and grow the company (like increasing the no. of chairs, buying a nearby builing and expanding etc.)

Free cash flow to the firm is essentially money flowing to the owners of the company - in this case, the owner of the company.  In the first case, free cash flow to the firm is equal to the net income. In the second case, it is again the net income - This is because the firm basically belongs to the owner and hence it makes no difference whether the money went to the firm or the owner. In the last case, it is 0. The idea here is that the reinvestment will create new assets and hence more cashflows in the future than otherwise. In reality, short of slowly running the company into the ground, there is no escaping from reinvesting: Every asset depreciates - the cushions in those chairs wear off, the vacuum cleaner and hair blower die slowly, the building slowly deteriorates and so on. Even if the owner isn't reinvesting, at least part of the cash flow, into creating new assets, he will have to reinvest to make up for the depreciation. For instance, at some point, the chair depreciate to scrap and he has to buy new chairs. Having said that, there is no point is reinvesting all the net income into the company forever. After some point, the owner has to benefit from all the reinvestments - he has to take out dividends. In fact, the economy itself may not allow continuous reinvestment and expansion - Even if the owner keeps opening more and more shops from the cashflow coming from existing ones, at some point, he will run out of hair to cut. 

So, in the realistic case, free cash flow to the firm is whatever is left of net income after some of it is reinvested. If we want to evaluate this business, we can use the most recent free cash flow thus computed into equation $\eqref{eq:PV}$. Of course, we have lot of unknowns to compute before we can work out the equation. But before we figure out other unknowns, there is an important nuance about cashflows: We have considered a simple buinsess where there was only one equity owner and no creditors. But many businesses get loans to partially finance the assets. In such cases, we can buy the entire company (all its assets) or just the equity. To understand this, think about buying a house: In the case where the counterparty had partially financed the house originally using a loan and is still holding some of it, you are buying the entire asset by paying some money to the equity owner and the rest to the bank. This is equivalent to buying the entire firm. Suppose, instead, you pay off just the equity owner his portion of the house value, and simply transfer the loan from his name to your name. This is equivalent to equivalent to just buying the equity in a company. When we buy stocks in a company, this is what we are buying - just (part of) the equity.

Since buying stocks is what we are interested in, we need to calculate the value of the equity in the firm - We can either directly value the equity or value the entire firm, and then, at the end, subract out the liabilities to get the value of the equity. Both methods should theoretically yeild the same result. However, calculating free cashflows are different between the two approaches.

+ In the earlier example of a barber shop, there was only one owner and no creditors. In this case, value of the equity and the value of the entire firm are one and the same. So the cashflows used was net income minus reinvestment is appropriate for valuing the firm as well as just the equity
+ But if the company has creditors, when valuing the entire firm, one cannot use net income as the base. This is because, if you buy the entire company, then it means you paid off the equity owner as well as the creditors. So although the company had interest payments going on before the purchase, there won't be any interests to pay after the purchase: It would be misleading to use net income as the base for valuation as it came after interest payments (and taxation). After the purchase, assuming the return on invested capital remained the same, the net income would resemble latest EBIT after tax is paid on it, i.e., EBIT(1-tax). Hence it makes more sense to use the most recent EBIT(1-tax) as the basis for the valuation. 
    + Note that EBIT contains, apart from operating income, extraordinary income aand other gains. So, while considering value of the company, operating income (with interests paid on it) may be a better indicator than EBIT
    + Operating Income (1-tax) is called "after-tax operating income"
+ If you want to evaluate the equity directly, then it does makes sense to use the latest net income as the basis for evaluation as this is the cash flow to you after you buy the equity, assuming ROIC remained the same.
+ When evaluating the entire firm (with EBIT(1-tax) as basis), the free cash flow is called FCFF - Free Cash Flow to the Firm. So the $FCF$ in equation $\eqref{eq:PV}$ is replaced by FCFF, and the precent value can be called $PVF$ - Present Value of the Firm. Similarly if you are valuing just the equity, then the $FCF$ should be replaced with $FCFE$ and $PV$ renamed $PVE$

### Nuances about free cashflows
So far, our take on free cashflows is that it is either after-tax operating income or net income with reinvestment needs factored in. But we didn't mention how exactly to go about figuring out how much the firm actually reinvests - In other words, how to come up with free cashflows by looking at the fundamentals. There are some problems associated with that. Part of the issue is that accountants, while coming up with operating or net income factor in or out non-cashflows for accounting reasons which mask the true value of the operating or net income. Particularly,
+ They factor in depreciation and amortisation of capital assets created in earlier periods for the purpose of getting tax relief on them. But these aren't actual cashflows out of the business during that period. So we need to add D&A expense back to our original cashflow calculation
+ They do not show capital expenses made during the period in question as they amortise it in later periods. But these are actual cashflows going out. So they need to be subtracted from our original cashflow calculation. That also covers the reinvestment needs
    + While considering capex, one should only consider capex to create operating assets. Capex spent on creating non-operating assets are basically like cash in the balance sheet, but in a different form. 
+ They use accrual accounting - It means they record, as "accounts receivable" in the balance sheet sale amounts right after the sale is done. They don't wait until the cash actually comes into the company. The other way around is also done - whenever purchases are made from vendors, they add the purchase amount to "accounts payable". Again, the cash hasn't yet gone out of the company. Operating and net income are calculated as if the moneys mentioned have already realised. To correct this, we need to subtract out accounts receivable from, add accounts payable to after-tax operating income or net income. The extra nuance here is that, there is a chance some of the accounts receivable/payable from earlier period may have been carried over. So we first need to net out the carry over items. 
+ If inventory increased compared to the last period, then cash went out of the company - but it doesn't appear in the income statement. So change in inventory must be subtracted from
+ Working capital, defined as current assets minus current liabilities. Current assets include inventory and accounts receivable. Current liabilities include accounts payable. So, many just subtract out change in working capital from after-tax operating income/net income. But this may not be the right thing to do because:
    1. Current assets also includes cash. 
    2. Current liabilities already include interest on debt, which we have already factored in
+ Professor Damodaran uses the term change in "non-cash, non-debt working capital" in place of just the change in working capital to stress this point.
+ While evaulating FCFE alone, principal payments made to creditors need to be subtracted out as these are cash flowing out of the equity owners. If principal payments are subtracted out, then new debts issued must be added to the calculation. 
+ Considering all this, the FCFF and FCFE calculations are given as:

$$
\begin{align}
FCFE &= \text{Net Income} - \text{CapEx} + \text{Depreciation} - \text{Change in non-cash WC} + \text{New debt issued} - \text{Debt repayments}\\
     &= \text{Net Income} - \overbrace{\left[\text{(CapEx - Depreciation) + Change in non-cash WC - (New Debt issued - Debt repayed)}\right]}^{\text{Equity reinvested}}
\end{align}
$$

and, 

$$
\begin{align}
FCFF &= \text{Op Income}(1-t) - \text{CapEx} + \text{Depreciation} - \text{Change in non-cash WC} \\
     &= \text{Op Income}(1-t) - \underbrace{\left[ \text{(CapEx - Depreciation) + Change in non-cash WC}\right]}_{\text{Earnings reinvested}} 
\end{align}
$$

+ The reinvestment rates (rir used in PV calculations) are given, while valuing equity, as:

$$
\text{Equity reinvestment rate (rir)} = \dfrac{\text{Equity reinvested}}{\text{Net Income}}
$$

and while valuing the firm as,

$$
\text{rir} = \dfrac{\text{Earnings reinvested}}{\text{Operating Income}(1-t)}
$$

Note: In the professor's equations, he uses EBIT(1-t) in place of operating income(1-t) in his equations. This is probably because he is looking at the cleaned up fundamentals: He is cleaning up for R&D, leases as mentioned earlier, and normalises earnings. "Normalising" means taking into account the possibility, in case of cyclical companies or black swan events, that the most recent fundamentals may not represent typical fundamentals and using instead a normalised - i.e., smoothened version that is more typical. Also, for tax, in these calculations, [the professor suggests](https://pages.stern.nyu.edu/~adamodar/New_Home_Page/valquestions/taxrate.htm) using either marginal tax rate for all the future cash flows or effective tax rate for just the first few future cashflows, then using marginal tax rate thereafter. Basically, the central theme here is that a company's present figures - fundamentals, tax rate etc. - may not be truly representative of its future. 

All the above are applicable to net income in his formulae as well. 

An important point about free cashflow based evaluation is that it can be used only to value operating equity, if using FCFE, or operating assets, while using FCFF. So at every step of the process, we need to ensure that non-operating assets aren't accidentally counted in. Once the operating assets' cumulative value is calculated, the value of non-operating assets (determined by using some other method than DCF, Ex. Fair market value) need to be added at the end. See the section 'Guidelines for practicing valuation' for more details

Since we count out non-operating assets, they have to be valued differently - say at their market price today - and added to the $PV$ coming out of DCF. For ex., if the firm has cash, it can simply be added. If it has government securities, the market value of those government securities can be added. If it is a fixed asset - like a building - then one has to figure out what the market value of that building is using real estate data.

## Estimating growth
While valuing the firm, we want to find out the expected growth of after-tax operating income, and, while valuing the equity, the growth in net income. Growth comes from how much is reinvested, and how much return do we get out of that reinvestment.

For firm valuation:
$$
g = rir(firm) * ROC \\
ROC = \dfrac{\text{Operating Income} (1-t)}{\text{BV of debt}+\text{BV of equity}}
$$

For equity valuation
$$
g = rir(equity) * ROE \\
ROE = \dfrac{\text{Net Income}}{\text{BV of equity}}
$$

+ The operating and net incomes above are corrected (for misrepresentations involving R&D, leases etc.) and normalised figures
+ Book value of capital/equity is used instead of market value because book value represents the actual amount that is invested
+ ROC/ROE are considered constants for the phase for which growth is estimated. This needs to be carefully evaluated before use
+ For someone investing in a company, ROC/ROE formula above gives the correct figure for comparing the investment's return with that of a different investment, say, an fixed deposit.  If we want to think of the company as growing at, say, 5% on a fixed capital of, say, a \\$100 part of whose returns are always reivested, in the first year, the company's returns, i.e., its income, would be \\$5. Say \\$2 from this earning is reinvested into the company, then the next year, the earnings would be \\$ $102*0.05$. If we divided this value by previous years accumulated capital of \\$102, then we will get the correct return percentage of 5%. So it makes sense to take any year's income and divide it by the (accumulated) invested capital as of the year before to find out the return percentage. 

+ If the firm has non-operating assets, then they need to be kept out of return on equity/assets. So in the numerator, income from non-operating assets must be netted out. And in the denominator, BV of non-operating assets must be netted out.

## Estimating the cost of capital
+ Since capital can be equity or debt, we need to calculated both and combine them.  
+ For calculating the costs of equity and debt, we should use latest figures derived from the market. The fundamental point here is that "cost" is what it takes you to buy something, while "value" is its actual worth - This is what we have established in the upper half of this document. What does it cost you to buy equity or debt of that company? It is whatever the market asks you to pay for them today. 
+ For cost of equity:
  + One could use CAPM with $\beta$ directly derived by looking at the correlation between the market and the company stock. But the professor argues that since the sample size is too small, such a $\beta$ would be "rotten". So he chooses to calculate, what he calls, the bottom up $\beta$ for the company: He calculates $\beta$ for the entire industry, unlevers that $\beta$ with average industry leverage and the relevers it using the leverage of the firm in question. Additionally, he "cleans up" the $\beta$ for cash. The idea is that, the $\beta$ for the industry is not rotten because of the large sample size and is a true proxy for risk. Then, within that industry, a company that uses more leverage than the average leverage is more risky. The company with more cash is less risky and so on.
  + In many of the professor's spread sheets, he also mentions a "total beta", which is the true beta for an industry for an undiversified investor - This, he says is useful while evaluating investments into a startup. In fact, for retail investors who are undiversified, this figure could be considered more appropriate as the basis for calculating the bottom up beta.
  + If a company does business globally, then either a global beta for that industry can be used for bottom-up beta calculations or a beta for that industry in every country could be scaled up using the country risk and weighted by the amount of revenue to the company from that country and summed up to arrive at the industry beta. Alternatively, one could directly use the global beta for that industry as a shortcut.
+ For cost of debt:
  + Use cost of debt for long-term debt and not short-term debt. 
  + If the company has been given a rating by S&P or Moody's, you could use the the default spread for that rating. Add that spread to the risk free rate and you are done.
  + If bonds of the company are publicly traded, you can check the yield on their long-term straight bond directly (Straight bond is one without embedded fancy options). However there is a chance that a risky company may have carved out low risk assets as collateral to release a low-risk bond. If you use that figure, then you will be underestimating the cost of debt.  
  + If none of these methods are suitable, calculate a bottom-up default spread by looking at the interest coverage ratio (averaged over several periods) and use a look-up table between interest coverage ratios and default spreads (from any rating agency). Add this to the risk free rate to come up with the cost of capital.
  + If the company chiefly operates in an outside country, one needs to add the country risk factor weighted by how much business the company does in that country.
+ While evaluating the entire firm, the weighted average cost of capital (WACC) is calculated from cost of equity and cost of debt using weights based on the market values of the debt and equities in that firm. Calculating the market value of the equity is somewhat straightforward - "somewhat" because there could still be two or more classes of shares and some may not be traded on the securities exchange (or if is a private company for the same reason its stocks aren't traded). For calculating the market value of the debt, professor suggests calculating the PV of an imaginary bond: Use the book value of the debt as par value of that bond, use the latest interest expense as the yearly coupon payment, use the cost of debt calculated earlier as the discount rate, and use the "weighted average maturity of the debts" (By that, I think he simply means weighting each debt's maturity by its proportion of the total debt)
+ One should discount the weight on the debt while calculating WACC with the marginal tax rate because of the tax benefits financing an asset with debt accords. Note that we should use the marginal and not the effective tax rate because the interest expenses save tax on the margin. 

[Prof's lecture on cost of capital](https://www.youtube.com/watch?v=N_FH89DCdGs&list=PLUkh9m2BorqnKWu0g5ZUps_CbQ-JGtbI9&index=6)


## Other factors for valuation
+ Some companies (ex. Google, Linkedin, UPS) have two classes of shares: Usually one of the classes has vastly higher (ex. 10x) voting rights than the other class. Professor says that the share with higher voting right should be [valued at a premuim](https://aswathdamodaran.blogspot.com/2011/05/dual-share-structure-google-model.html). One shortcut, IMO, if both classes are traded on national securities exchanges or OTCMK, is to check how much premium the market is attaching to the higher class and use that premium factor. 
+ Throughout the "Valuation by Damodaran" book, there are inconsistencies when it comes to keeping out non-operating assets while valuing operating assets - sometimes, he nets out only cash, while at others, cash and marketable securities, and in a video on youtube, while talking about reinvestment, he mentions that only reinvestments into operating assets should be considered. The overall idea is that, while valuing a firm, non-operating assets and their incomes need to be kept out of the calculations until we have finished evaluating the operating assets. 
+ Professor asserts that $g2$ cannot be faster than the economy it is operating in. But I think it is more like $g2$ cannot grow faster than the market the company is serving. Most often this market, if it is mature, may not grow faster than the economy. So in some ways the professor is giving us a thumb rule.
+ Compared to phase 1 of $PV$ calculation, almost all fundamentals in phase 2 will be different. This is because we are assuming that, in phase 1, the company is a growing company, and in phase 2, it is a mature company. A stable growth company, as compared to a growing company,
    + Less risky. Consequently both is cost of equity and cost of debt should be less. Cost of debt can be determined by looking at higher grade bonds, say, AAA, or by looking at the cost of debt of companies that are considered stable growth companies today. Cost of equity can be set to be equal to market risk. i.e., $\beta=1$, and correspondingly, $ROE$ to expected market return
    + Use more debt. Cost of capital is lesser due to more debt compared to equity (in addition to cheaper debt and equity)
    + Reinvest less
+ Period of extraordinary growth, that is, $N$ in our $PV$ calculations, comes just from good judgement. Professor doesn't propose any methodical way of arriving at it
+ The growth rate $g2$ during the stable growth period can be decided by looking at the growth rate of a stable-growth company today, in the same industry, that our candidate company will resemble when the extraordinary growth period is over. 
+ $\beta$ can be set to 1 or close to that for the stable growth period resulting in cost of equity looking like equity risk premium of the industry. The expectation is that mature companies in the same industry today have their cost of equity similar to ERP
+ During the stable growth period, the return on equity is not expected to be much higher than the cost of equity. 
+ Reinvestment rate during the stable growth period $rir2$ is usually reverse engineered from the ROE and $g2$
+ Risk-free rate used in a calculation should match the intended investment period. If we are valuing a stock with the intention of investing in it for 10 years, use 10 year T-bond as $R_f$ in CAPM calculations, and not a one-year Treasury note rate.
+ The general method used for valuing companies as discussed so far takes into account, directly or indirectly, the wisdom of the management's strategies, effect of capex on R&D and tangible parameters as book and market values of equity and debt, even the hope and promise people see in the economy (by using ERP, RFR). However the market value of the assets is not reflected. For example, firm may have assets such as real estate or metals or oil whose book value could mislead us since, a. the market value may have charged a lot b. the depreciation may have wiped out value for accounting purposes while it may not actually make any real-world sense. Peter Lynch refers to such companies as "asset-play"s. Going one step further, a company may be holding significant shares of another company whose true value - in the same manner as described above - isn't reflected by its books. No generic python code or valuation method can be applied in such situations.  
+ Cyclicals rise and fall with the economy. If we use the valuation method as described in this document, it could result in severe miscalculation

# Guidelines for practicing valuation
The general idea is that we evaluate the actively managed operating assets and rest of the assets separately. Operating assets are assessed through fundamental analysis and has implications on how successful the companies business plan is, and how well its managers execute. Rest of the assets including cash and marketable securities, non-operating assets and minority holdings in other companies. One can take them at their market value and add their value at the end to the value of the actively managed operating assets. In the case of minority holdings, one could conceivably also evaluate the company held using fundamental analysis and calculate the portion of equity owned by our company. Income generated by these non-actively-managed assets is not considered anywhere in the various formulae used while valuing the actively managed operating assets. Because we are evaluating these non-actively-managed assets separately, and often, *differently* (not using cashflows), these cashflows aren't of much relevance. 

+ Items to be included under 'Cash and marketable securities':
    + Cash and cash equivalents
    + Short-term financial instruments
    + Current portion of long-term investment securities
    + Note that all the above items are from 'Current assets'. Investments in long-term securities are not included as they are considered to be operational, growing on the basis of the managment's decisions.
+ Items under 'Interest expenses':
    + Interest expenses
    + Interest expenses on lease liabilities
+ Equity in non-controlling interests is **not** included as of the 'equity' in valuation formulae. It is later added on to the value of the operating part of equity to get the total value of equity (One can use MV of the non-controlling interest or do DCF valuation of that other company as well and then take their minority holding percentage of the result of that valuation).
    + The reason seems to be that, like cash, non-controlling interests aren't actively managed by the company, and unlike long-term securities whose purchase and sale are actively managed by the company
+ The following are considered as debt:
    + From current liabilities: 
        + Short-term borrowings
        + Current portion of long term debt
        + Current portion of lease liabilities
    + From non-current liabilitie:
        + Bonds payable and long-term borrowings
        + Lease libilities

## Steps
  1. Come up with a financial balance sheet from accounting balance sheet by capitalizing R&D and Lease commitments
  2. For computing future cashflows during the extraordinary growth phase, use the operating income, growth rate and rir and tax rate: Use effective tax rate for the first few years and then move on to marginal tax rate. (Instead of calculating FCFF and using it for every year with $(1-rir)$ and $(1+g)$ substitute FCFF with $opincome*(1-effective\ tax\ rate)$ for the first few years but substitute FCFF with $opincome*(1-marginal\ tax\ rate)$ for later years. "First few years" is a discretionary choice. Effective tax rate is mentioned in 10K. Marginal tax rate is available [here](https://pages.stern.nyu.edu/~adamodar/). Extraordinary growth phase should be less than 10 years long. Preferably 5 years.
  3. Calculate the MV of debt find the yield on the companies long-term striaght bond. Use [FINRA search](https://finra-markets.morningstar.com/BondCenter/Default.jsp). Use the BV of debt without leases from the financial balance sheet as the face value of an imaginary bond, interest expense as the coupons and average maturity time of the debt (find from 10K) as the maturity term. And discout these by the yield of the straight bond. Add back the lease amount to get the MV of debt.
  4. For RFR, use the latest [30Y Treasury bond yield](https://finra-markets.morningstar.com/BondCenter/Default.jsp). Most of the Professor's other data using data that far out. So it is safer that using the 10Y treasury bond
  5. For calculating cost of debt, add to RFR the default spread of company's bond if available - This is basically similar to the yield of the straight bond minus RFR. If the company does a lot of business in a country outside USA, add a scaled version of the country default spread to the above sum. The scale is a discretionary choice - one possibility is to use the proportion of revenue coming from that country. Use this method for as many countries as required. Rating agencies give countries ratings - You can find the default spread for the rating using look up tables which are also given by rating agencies. If you have difficulty finding a straight bond, or you think the straight bond isn't representative (because bad companies could carve out good assets to back up a bond), use [synthetic rating](https://pages.stern.nyu.edu/~adamodar/) based on the interest coverage ratio
  6. For calculating beta,use [U.S. industry betas](https://pages.stern.nyu.edu/~adamodar/pc/datasets/betas.xls).
      - If you decided to use Unlevered beta, relever by multiplying it with $\left(1+(1-t)\dfrac{D}{E}\right)$, where the tax is marginal tax and weights $D$ and $E$ are market value of debt and equity respectively. Use this relevered beta as the firm's beta
      - If you decided to use Unlevered beta corrected for cash, then multiply it with $\left(1-\dfrac{cash}{D+E}\right)$ and then by $\left(1+(1-t)\dfrac{D}{E}\right)$. The third option is to use the average Unlevered beta corrected for cash (last column in the [U.S. industry betas](https://pages.stern.nyu.edu/~adamodar/pc/datasets/betas.xls)
      - If the company operates across multiple industries, multiply its revenues from each industry with their corresponding [EV/Sales number](https://pages.stern.nyu.edu/~adamodar/pc/datasets/psdata.xls) to get a better idea of how much the market values the revenue in each industry. Use the proportions of resulting scaled revenues as the basis for calculating weights with which to scale the unlevered betas of the corresponding industries. Use the result as the unlevered beta for finding firm beta. Again, you can use unlevered beta or unlevered beta corrected for cash.
  7. Use implied ERP from Professor's [home page](https://pages.stern.nyu.edu/~adamodar/) - the first figure against "Implied ERP as on lastest_date". Then calculate the cost of equity as $RFR+\beta*ERP$.
      - If the company does business in an country outside USA as well, then use $RFR + \beta*ERP + \lambda*CRP$, where CRP is [country risk premium](https://pages.stern.nyu.edu/~adamodar/) and $\lambda$ is the relative exposure of the company to this country. $\lambda$ is a discretionary choice. Check how the professor calculates it for Embraer in his book, Illustration 2.4. One could compare the proportion of revenue of this company in that country with that of an average company: So if this company makes 10% of its revenue from India while the average company in its industry does 5%, then use $\lambda = \dfrac{10\%-5\%}{5\%} = 1$. That way if your firm does same amount of business in India as the average company, the $\lambda$ would be 0, which is appropriate. In fact, going further, if your company is less exposed to India than the average company, it actually makes sense to have a negative $\lambda$. 
      - You can do this for as many countries as you want. If your company does business in diverse geographies, consider using the region risk premiums instead of CRP. They are only available in [fcffginzu.xls](https://pages.stern.nyu.edu/~adamodar/pc/fcffginzu.xlsx)
  8. Claculated Wacc. Use marginal tax rate.
  9. Calculate PV of extraordinary growth phase. 
  10. For terminal value, use ROIC equal to the risk free rate. 
  11. Add PV of extraordinary growth phase with the terminal value to arrive at the value of operating assets.
  11. To arrive at the value of equity, after operating assets are valued and added to non-operating assets (cash and marketable securities), subtract from the result the market value of the liabilities . The market value is appropriate here because to buy the whole firm, you would have to pay off the creditors the market value of debt. Think of it this way: If you want to buy the firm, you need to buy up all the outstanding bonds, which means paying the market value for them. The topic "Calculating the cost of capital" specifies the technique for converting book value of debt to its market value. Note that it seems the professor calculates the market value of debt before adjusting it for lease liabilities and then simply adds lease liabilities to come up with the market value of all the libilities. 

## Formulae
### For firm valuation
$$
\text{Reinvestment rate} = \dfrac{\text{(Capex - Depreciation)} + \text{Change in non-cash non-debt WC}}{\text{Operating income}*(1-t)}
$$

$$
\text{ROIC} = \dfrac{\text{Operating income}*(1-t)}{(\text{BV of Equity} + \text{BV of debt} - \text{Value of non-actively managed assets})_{prev\ year}}
$$

$$
\text{WACC} = \text{Cost of equity}\left( \dfrac{E}{D+E} \right) + \text{Cost of debt}*(1-t)\left( \dfrac{D}{D+E} \right)
$$

where, $E$ and $D$ are the market values of equity and debt respectively

$$
\begin{align}
\text{Value of equity} = &\text{Value of operating assets} \\
                         &+\text{Value of non-actively managed assets} \\
                         &-\text{Market value of debt} \\
                         &-\text{Value of minority interest in consolidated companies} \\
                         &-\text{Unfunded health care or pension obligations} \\
                         &-\text{Expected litigation payouts}
\end{align}
$$

### For equity valuation
$$
\text{Reinvestment rate} = \dfrac{\text{(Capex - Depreciation)} + \text{Change in non-cash non-debt WC} - \text{(New debt issues - repayments)}}{\text{Operating net income}}
$$

where,
$$
\text{Operating net income} = \text{Net income} - (1-t)*(\text{income from non-actively managed assets})
$$

$$
\text{ROE} = \dfrac{\text{Operating net income}}{(\text{BV of Equity} - \text{Value of non-actively managed assets})_{prev\ year}}
$$

# Valuation pitfalls and general notes
+ Number of outstanding shares is the single most important figure while deciding whether or not to buy a stock. So don't rely on AlphaVantage or Yahoo finance. Check the 10K/Q yourself and make sure this figure is right. For LEVI I got screwed because both the aforementioned service quoted the number of outstanding shares as one-fourth of what it actually was!
+ The growth figure you come up with for operating income shouldn't be higher than the growth than what the company sets out in its 'Business Strategy' section in the 10K. You cannot accord them more growth that they themselves expect
+ Sometimes companies mention their effective tax rates for various years in their 10K. If it has been steady, you can use that figure. Otherwise, you can take the arithmentic average of the last few years.
+ While valuing a company as on a past date, if there have been stock splits since then and now. make sure to recompute EPS and DPS and use post-split value of outstanding shares (Affects MV of equity)
+ If calculating historical growth use net income figures and not EPS or DPS, because EPS and DPS are affected by stock issues, buybacks and splits and may give a wrong impression
+ If using TTM (Trailing Twelve Months) for evaluating a stock in the middle of the year, check to make sure you aren't in that sweetspot time when last years dividends don't show up in TTM and this year's dividends haven't been declared yet. You will end up grossly overvaluing the company if you don't check this
+ At the end of an intrinsic valuation, it is a good idea to look at implied relative valuation measures (such as value to earnings ratio) and compare that P/E. This is to make sure you haven't made a gross error during intrinsic valuation. For example, if the P/E is 12 and your valuation implies it should be 57 and the industry average P/E is 15, then it probably means you screwed up some aspect of the intrinsic valuation.
    + It may also be a good idea to evaluate a company right after 10-K before you do a TTM based valuation. The idea is that, you will not miss dividends like mentioned in the previous paragraph. We can then do a TTM based valuation and compare - since the max difference between a 10-K and TTM can be 11 months, there shouldn't be a large difference in valuation results. If they are off by too much, you can go back and investigate your calculations further.
+ If a company's credit rating is speculative grade, then you have to seriously consider the possibility of equity value in the event of a default. Usually it is zero. So you should think of a risk premium for bearing this risk and require a cheaper stock price.
+ Cashflow statements in 10-Qs may not be for three months. So while doing TTM based valuation, the only way to calculate reinvestment items (capex, Depreciation, Change in WC etc.) may be to look at the difference between two balance sheets at the beginning and the end of the period

# General thoughts on valuation
Before investing in a company, you need to have both the numbers and the story to back up your decision to invest. Numbers part is obvious - that is the valuation. The story part is what backs up the discretionary choice we made during the valuation, such as how many years will a high growth company take to become a stable growth one, how much will be the stable growth rate etc. For instance, if you are investing in a real-estate company, you can say that demand for housing is very high in the country and the supply isn't there to meet the demand - So you think in the near term, the company in question is going to benefit. Basically the story is your version of that company's business plan - Your story basically consists of their USP, market size etc.

Beyond the business plan, one should also take into account the word in the street about the company's products. In today's information age there is no excuse for not doing this. Many businesses make products such as cameras, mobile homes etc., whose users talk about their products on forums. Reading forum posts should be part of your market research before investing in a company. These are the places where you can find, as Rober Shiller would put it, if the narrative virus is spreading. 

When we think of a starting a company, we want our company to have all the following characteristics:
  + Great income statement
    + Rapid revenue growth in a market that is itself expanding
    + Great margins
    + Low marginal costs
    + High scalability
    + Great interest coverage ratio
  + Great balance sheet
    + Enough cash buffer for deal with emergencies or unforeseen losses
    + Fast moving inventory
    + Assets, if PPE (Property, plant and equipment), hold market value for a long time

These are indeed qualities any good company must possess. For example, even if the company' earnings in absolute dollars have historically been very good, if margins are poor, a slight change in the market, such as the entry of a competitor or a slight change in the purchasing power of the consumer, could wipe out those great earnings would be wiped out quickly. On the other hand, if the margins are high, if a competitor emerges the company could lower the price of its product for a short term at such a level that the competitor can't afford to compete. 

It is important to remember that valuation is less a sciece and more an art: The growth projections do not (and cannot) take into account changes in management style of the company or changes in management itself, the general state of the economy, future innovation that can make the company's products/services obsolete and many more factors. It also doesn't take into account the possibility of the company constantly raising a whole lot equity, or issuing a whole lot of debt as its primary form of investment instead of reinvesting its own income. Then there is the fact that growth projections for cyclical companies could be off so much that the we may end up valuing the company 4x it's actual value! The point of the valuation is not so much coming up with an exact figure upon which to make investment decisions, as evaluating the success of the company's business plan and the merits of its managers 

One could choose to start valuation by selecting companies at the forefront of innovation in a new, expanding market, or on could choose to use a process of elimination to arrive at companies to evaluate via a process. The former will have personal bias built into future investments at the very beginning of the valuation process itself, while the later may take too much time to work through.

If a process of elimination is used, one can quickly eliminate
+ Young companies that haven't found their scalable business model
+ Stable companies that have lost track of their business model, trying all kinds of new things (Let us get into IoT because it is so hot now) - Peter Lynch calls this "Diworsification"
+ Companies that generally are in bad health - bad interest coverage ratios, no emergency cash, very bad margins, too much leverage, using loans to pay dividends etc.
+ Companies that are all over the place - like conglomerates - that have their profit spread out in multiple subsidieries in somewhat even fashion (In other words, conglomerates like Alphabet where the lion's share of profit comes from a single subsidiery, namely Google, may be OK).
+ Any business that doesn't have a simple, straight forwards story. One should be easily able to create a business canvas for that company.
    + There may be businesses with complex stories that may be undervalued. But tracking them, after investing in them, and understanding whether they are performing as expected may be too much work. 
+ Companies where accountants seem to be playing too many tricks, and companies where the management desperately tries to maintain a good image (by, say, smoothening out their dividends), while the reality isn't that good. We can call these "Modi companies" - basically companies that are trying to deceive investors

The general idea is that an individual investor is spoilt for choice - there are thousands of companies in a chosen county and there are several countries to choose as well. But at EoD, the individual investor is only going to invest in not more than 10 companies. So eliminating companies with the most stringent of criterion will still result in too many choices - one shouldn't worry about using too many rules or eliminating most companies quickly. 

Young growth companies may not have a positive or steady (or both) operating margins. In that case, we should start with the revenue, make some assumption about operating margins, use sales-to-capital as a measure of returns, and do an FCFF valuation. FCFF method is probably more suitable than FCFE method as FCFF nethod would require less assumptions - particularly regarding amout of leverage the company would use in the future.

Real growth in revenues (as opposed to growth because of increasing price of its produts) may come about for a company because they may be expanding their market share, or because the market itself is expanding. A company that is undervalued, but growing, serving a market who size is expanding probably make for the best choices for investing.

Companies that have a brand recognition shouldn't be investing too many dollars to generate one sales dollar. 

When we label a company as a stable growth company or young company with great potential etc, we should verify our story from various angles. For instance, a stable growth company should have the following characteristics:
+ It must be spending very little on R&D
+ It should have higher leverage ratio than young growth companies in the same industry
+ Its retention ratio must be small compared to young growth companies in the same industry
+ Its revenue growth must be steady and below the growth rate of the economy, margins steady save black swan events and its dividends must be regular and smooth

The proof of the pie is in the pudding - There may be many qualitative drivers of value, but a companies fundamentals are the only clear evidence that those qualitative drivers are indeed working. So it is OK to say that a company is innovating, but if they aren't able to monetize their innovation, it is of no use to the investor (Ex. Google glass)

It may be a good idea to do FCFF instead of FCFE based valuation even if the net income figures look stable and predictable. The reason is, when doing FCFF, you will be forced to look at the bonds issued by the company, their credit rating etc. You may realise that bad credit rating isn't easily visible in the financial statements. If the company cashflows do not meet its obligations at some point in the future, creditors may foreclose it and the equity may become worthless.

COGS may give an idea on whether increase in revenue compared to previous period is due to increasing volume of sales or increasing price of units. Actually some companies specify what was the reason for increase in sales.

Companies where the management doesn't own a large share of the stocks should sell at a discount companred to companies where the management owns a large share of the stocks. The reason is, managers' compensations are usually linked to the performance of the stock in the stock market, and if they aren't holding a lot of stocks, they may make decisions that is good for the stock price in the short term, but bad for the company in the long term. Having said that, public companies whose holdings are extremely concentrated in the hands of insiders could see its stock manipulated by insiders (Ex. Adani Enterprises). Sometimes the "promoters" might transfer money from their privately held companies to their public company to shore up the books. Sometimes they pump up the stock prices of their public company with bad balance sheet, then use the pumped up shares as a currency to buy companies with good balance sheets and then sell off assets from the aquired company or transfer assets to the parent. They may also use the acquired company's stocks as a collateral to get loans for the parent company.

# Quirks
+ If investors think a company is going to be acquired, they may send bond yields negative as they trip over each other to buy the bonds - The idea is that, the acquiring company will have to buy off those bonds at market price, and so the investors will get paid - So the negative yields should not be looked at using conventional ideas about inflation. opportunity cost etc. - Prevailing macroeconomic sentinment may be one of inflation, and bond markets in general may be seeing positive yields, and still, this company in question will be seeing negative yields, as yields, at that point, have no meaning for that bond. 

# Musings about value
+ The value of anything is the value of another thing that you would exchange, directly or indirectly, to buy that thing. Think about barter trading: The value of a bag of rice is equal to the value of, say, a piece of furniture. There is no alternative way of thinking about value of an item. It is like how, in physics, position and time are always relative to some reference point. 
    + If you don't keep a maid, do all the house cleaning yourself, the value of the house cleaning tasks you do is equal to the value of the other task that you could be doing, but which you swapped for cleaning the house instead. For ex., if it takes you 10 hours to clean your house, but if you didn't do it, you could have, say, done 10 hours of some consultancy work, then the value of (or the cost of) the house work, in INR, is how much you would have earned doing consultancy work instead. This is what opportunity cost is. "Opportunity cost" is fancy way of describing basic barter trading. Now if the opportunity cost is, say, Rs. 10K, and a maid could be employed to do the same work for Rs. 2K, then by employing a maid, you have reduced the cost of house cleaning by 8000 rupees! In other words, there is no intrinsic cost of cleaning the house - it is only relational. In this example, it is rational for you to employ a maid because your opportunity cost is higher than the cost of keeping a maid. But why does the maid take up the job? It is because her opportunity cost of taking up this job is lower than Rs. 2K. That is, by accepting this job, she is forgoing another job that she could take, say, in a construction site - If that job would pay her less than Rs. 2K for the same 10 hour work, then it makes sense for her to accept your job offer
+ When you get a house loan for investing in real estate, the bank assesses the appropriate interest rate based on, among other things, the geographical location of the house and how much its value is expected to appreciate. Say, it is 10%. The question is, if the bank knows how to estimate the value appreciation a house in a locality, why doesn't it directly buy the house instead of lending you the money? That is because bank, as an organisation, has people who are well versed with looking at statistical data and coming up with decision like how much interest rate to charge. So what they are assessing is like the CAPM expected return or the risk (or vice versa). But you may be a domain expert in real-estate: You know what other people, including the bank don't know. For example, you may know that a certain type of interior decoration or a certain type of balcony view is psychologically satisfactory to house buyers - that is a knowledge you gathered from working in the industry for several years - something that bank and other people do not possess. This is like the $\alpha$ that fund managers deliver above CAPM required rate of return. In fact, one could expand this rationale for corporate loans as well: Why won't banks do oil exploration instead of giving loans to oil exploration companies etc.
    + Suppose you are the CFO of a company and you decide to take a bank loan, for say, 15% APR, for a project, then it means you think your companies domain extertise in the concerned project's domain, makes it less riskier, or more rewarding, that what the bank thinks. If this is not true, then you shouldn't be borrowing!
+ Stock buybacks will end up reducting equity value for equity owners who didn't sell their stocks back if the market price of the stock is higher than the actual value per share (of the equity). We can write:

$$
vps_{bbb} = stock\ price - premium
$$

where the subscript $bbb$ stands for "Before Buy Back". Suppose $N_{bbb}$ is the number of outstanding shares before buyback and $N_{abb}$ number of outstanding shares after the buy back, then we can write the value per share after buy back as:

$$
\begin{align}
vps_{abb} &= \dfrac{value_{bbb} - \text{Cash spent on buy back}}{N_{abb}} \\
          &= \dfrac{vps_{bbb}N_{bbb} - \text{Cash spent on buy back}}{N_{abb}} \\
\end{align}
$$

and the value lost per share as,

$$
\begin{align}
vps_{bbb} - vps_{abb} &= vps_{bbb} - \dfrac{vps_{bbb}N_{bbb} - \text{Cash spent on buy back}}{N_{abb}} \\
&= \dfrac{vps_{bbb}N_{abb} - vps_{bbb}N_{bbb} + \text{Cash spent on buy back}}{N_{abb}} \\
&= \dfrac{vps_{bbb}N_{abb} - vps_{bbb}N_{bbb} + (N_{bbb} - N_{abb})*stock\ price}{N_{abb}} \\
&= \dfrac{(N_{bbb} - N_{abb})*stock\ price - (N_{bbb} - N_{abb})*vps_{bbb}}{N_{abb}} \\
&= \dfrac{N_{bbb} - N_{abb}}{N_{abb}} * (stock\ price - vps_{bbb}) \\
\text{value lost per share} &= \left( \dfrac{N_{bbb}}{N_{abb}} - 1 \right) * premium
\end{align}
$$

Note that, in the above equation, $\dfrac{N_{bbb}}{N_{abb}}$ is greater than 1, and hence consequentially, value lost per share is greater than 0. If investors believe that the current stock price is perfectly justified, i.e., there is no premium, then there is no value per share lost and a stock buy back shouldn't affect the stock price. Conversely, when a company buys back its shares, since every managements job is to maximise value for their investors, the buy back can be viewed as reflecting the management's belief that the current stock price is lesser than or equal to the actual equity value of the company.
